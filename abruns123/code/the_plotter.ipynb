```python
import matplotlib.pyplot as plt
from distance_calc import diffm, reader, get_coords
from direction_of_motion import acc_reader, get_direction


lats,lons=reader("/workspaces/CP1-24-midterm/abruns123/data/circle_data/Location.csv")
distances=[]
xy_coords=[]
x_pos=[0]
y_pos=[0]
for i in range(1,len(lats)-1):
    distances.append(diffm(lats[i], lats[i+1], lons[i], lons[i+1]))

for i in range(len(distances)):
    xy_coords.append(get_coords(distances, distances[i], i))
    x,y=xy_coords[i]
    x_pos.append(x)
    y_pos.append(y)
plt.figure()
plt.plot(x_pos,y_pos)
plt.ylabel("Y (m)")
plt.xlabel("X (m)")
plt.title("GPS Position Map")
plt.grid()
plt.show()
plt.figure()
acc,times=acc_reader("/workspaces/CP1-24-midterm/abruns123/data/down_data/Raw Data.csv")
directions=get_direction(times, acc)
plt.plot(times,directions)
plt.title("Elevator Direction of Motion")
plt.ylabel("Z direction")
plt.xlabel("time (s)")
plt.grid()
plt.show()
```
